## Logistic Regression

### Evaluate the model on Convertible vs Minivan using the resized dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from dataloader import load_data
from cross_validation import CrossValidationDataset
from PCA import PCA
from logistic_regression import LogisticRegression
from utils import accuracy_01, plot

plt.rcParams["figure.figsize"] = (12, 8)


dataset = CrossValidationDataset(load_data('./resized/')[0], ['Convertible', 'Minivan'], 10)

train_loss_record = []
train_accuracy_record = []

holdout_loss_record = []
holdout_accuracy_record = []

for k, (train_data, train_labels, test_data, test_labels, holdout_data, holdout_labels) in enumerate(dataset.data()):
    n_components = 40

    pca = PCA(train_data, n_components)

    train_data = pca.apply(train_data)
    test_data = pca.apply(test_data)
    holdout_data = pca.apply(holdout_data)

    logistic = LogisticRegression(n_components)

    train_loss_record.append([])
    train_accuracy_record.append([])

    holdout_loss_record.append([])
    holdout_accuracy_record.append([])
    best_accuracy = 0
    best_w = logistic.w 
    for epoch in range(300):
        train_y = logistic.forward(train_data)
        train_loss = logistic.loss(train_y, train_labels)
        train_accuracy = accuracy_01(train_y, train_labels)
        train_loss_record[k].append(train_loss)
        train_accuracy_record[k].append(train_accuracy)
        logistic.backward(train_data, train_y, train_labels)
        
        
        holdout_y = logistic.forward(holdout_data)
        holdout_loss = logistic.loss(holdout_y, holdout_labels)
        holdout_accuracy = accuracy_01(holdout_y, holdout_labels)
        holdout_loss_record[k].append(holdout_loss)
        holdout_accuracy_record[k].append(holdout_accuracy)
        
        # record the parameters when best accuracy occurs
        if holdout_accuracy > best_accuracy:
            best_w = logistic.w
            best_accuracy = holdout_accuracy
        
    
    break

logistic.w = best_w
test_y = logistic.forward(test_data)
test_loss = logistic.loss(holdout_y, holdout_labels)
test_accuracy = accuracy_01(test_y, test_labels)
print('The test set loss and accuracy is {:2f} and {:.2f}'.format(test_loss, test_accuracy))


plot('loss', train_loss_record[0], holdout_loss_record[0])
plot('accuracy', train_accuracy_record[0], holdout_accuracy_record[0])

# plot the first 4 PCA component as required
plt.figure
plt.subplot(2,2,1)
plt.imshow(pca.top_eigen_vectors[:,0].reshape((200,300)))
plt.title('1st PCA component')
plt.subplot(2,2,2)
plt.imshow(pca.top_eigen_vectors[:,1].reshape((200,300)))
plt.title('2rd PCA component')
plt.subplot(2,2,3)
plt.imshow(pca.top_eigen_vectors[:,2].reshape((200,300)))
plt.title('3nd PCA component')
plt.subplot(2,2,4)
plt.imshow(pca.top_eigen_vectors[:,3].reshape((200,300)))
plt.title('4th PCA component')
plt.show()
```

```python
dataset = CrossValidationDataset(load_data('./aligned/')[0], ['Convertible', 'Minivan'], 10)

train_loss_record = []
train_accuracy_record = []

holdout_loss_record = []
holdout_accuracy_record = []

best_test_accuracy = 0
k_train_loss = np.zeros((10,300))
k_holdout_loss = np.zeros((10,300))
k_train_accuracy = np.zeros((10,300))
k_holdout_accuracy = np.zeros((10,300))
test_accuracy = 0
test_loss = 0
for k, (train_data, train_labels, test_data, test_labels, holdout_data, holdout_labels) in enumerate(dataset.data()):
    n_components = 40

    pca = PCA(train_data, n_components)

    train_data = pca.apply(train_data)
    test_data = pca.apply(test_data)
    holdout_data = pca.apply(holdout_data)

    logistic = LogisticRegression(n_components)

    train_loss_record.append([])
    train_accuracy_record.append([])

    holdout_loss_record.append([])
    holdout_accuracy_record.append([])
    best_holdout_accuracy = 0
    best_holdout_w = logistic.w 

    for epoch in range(300):
        train_y = logistic.forward(train_data)
        train_loss = logistic.loss(train_y, train_labels)
        train_accuracy = accuracy_01(train_y, train_labels)
        train_loss_record[k].append(train_loss)
        train_accuracy_record[k].append(train_accuracy)
        
        
        logistic.backward(train_data, train_y, train_labels)
        
        
        holdout_y = logistic.forward(holdout_data)
        holdout_loss = logistic.loss(holdout_y, holdout_labels)
        holdout_accuracy = accuracy_01(holdout_y, holdout_labels)
        holdout_loss_record[k].append(holdout_loss)
        holdout_accuracy_record[k].append(holdout_accuracy)
        
        k_train_loss[k, epoch] = train_loss
        k_holdout_loss[k, epoch] = holdout_loss
        k_train_accuracy[k, epoch] = train_accuracy
        k_holdout_accuracy[k, epoch] = holdout_accuracy
        
        
        # record the parameters when best accuracy occurs
        if holdout_accuracy > best_holdout_accuracy:
            best_holdout_w = logistic.w
            best_holdout_accuracy = holdout_accuracy
        
        
        
    # Average over all k trails
    logistic.w = best_holdout_w
    test_y = logistic.forward(test_data)
    test_loss += logistic.loss(test_y, test_labels) / 10
    test_accuracy += accuracy_01(test_y, test_labels) / 10
    
print('Test accuracy is {:.2f}%'.format(test_accuracy * 100))

# Average and std of loss and accuracy

plot('Average loss and std', train = np.mean(k_train_loss, axis = 0), holdout = np.mean(k_holdout_loss, axis = 0),\
    train_std = np.std(k_train_loss, axis = 0), holdout_std = np.std(k_holdout_loss, axis = 0))
plot('Average accuracy and std', train = np.mean(k_train_accuracy, axis = 0), holdout = np.mean(k_holdout_accuracy, axis = 0),\
    train_std = np.std(k_train_accuracy, axis = 0), holdout_std = np.std(k_holdout_accuracy, axis = 0))


# plot the first 4 PCA component as required
plt.figure
plt.subplot(2,2,1)
plt.imshow(pca.top_eigen_vectors[:,0].reshape((200,300)))
plt.title('1st PCA component')
plt.subplot(2,2,2)
plt.imshow(pca.top_eigen_vectors[:,1].reshape((200,300)))
plt.title('2rd PCA component')
plt.subplot(2,2,3)
plt.imshow(pca.top_eigen_vectors[:,2].reshape((200,300)))
plt.title('3nd PCA component')
plt.subplot(2,2,4)
plt.imshow(pca.top_eigen_vectors[:,3].reshape((200,300)))
plt.title('4th PCA component')
plt.show()
```
